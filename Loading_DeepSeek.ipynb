## Helper Function
This function is designed to start a background thread to run the command ollama serve using Python's subprocess module

```python
import threading
import subprocess
import time

def run_ollama_serve():
  subprocess.Popen(["ollama", "serve"])

thread = threading.Thread(target=run_ollama_serve)
thread.start()
time.sleep(5)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama.llms import OllamaLLM
from IPython.display import Markdown

template = """Question: {question}
Answer: Let's think step by step."""

prompt = ChatPromptTemplate.from_template(template)
model = OllamaLLM(model="deepseek-r1")
chain = prompt | model
```

```python
display(Markdown(chain.invoke({"question": "what is the value of x when we have this equation x+2=4"})))
```
